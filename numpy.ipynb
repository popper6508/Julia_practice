```python
import numpy as np
```

```python
from skimage.transform import radon, resize
```

```python
img = np.random.rand(256, 256)
```

```python
img
```

```python
f = resize(img, (128,128))
Rf = radon(f)
```

```python
import matplotlib.pyplot as plt
```

```python
import time
a = time.time()
plt.subplot(1,2,1)
plt.imshow(f)
b = time.time()
print(b-a)

a = time.time()
plt.subplot(1,2,2)
plt.imshow(Rf)
b = time.time()
print(b-a)
```

```python
np.save("f.npy", f)
np.save("Rf.npy", Rf)
```
