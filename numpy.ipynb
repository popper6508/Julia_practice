```python
import numpy as np
```

```python
from skimage.transform import radon, resize
```

```python
img = np.random.rand(256, 256)
```

```python
img
```

```python
f = resize(img, (128,128))
Rf = radon(f)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.subplot(1,2,1)
plt.imshow(f)
plt.subplot(1,2,2)
plt.imshow(Rf)
```

```python
np.save("f.npy", f)
np.save("Rf.npy", Rf)
```
